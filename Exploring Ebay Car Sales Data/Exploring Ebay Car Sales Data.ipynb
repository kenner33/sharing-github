## Cleaning Column Names

Let's convert the column names from camelcase to snakecase and reword some of the column names based on the data dictionary to be more descriptive.

```python
autos.columns
```

```python
autos = autos.copy()
autos.rename(columns = {'yearOfRegistration': 'registration_year', 
                        'monthOfRegistration': 'registration_month', 
                        'notRepairedDamage': 'unrepaired_damage',
                        'dateCreated': 'ad_created',
                        'dateCrawled': 'date_crawled',
                        'offerType': 'offer_type',
                        'vehicleType': 'vehicle_type',
                        'powerPS': 'power_PS',
                        'fuelType': 'fuel_type',
                        'nrOfPictures': 'nr_of_pictures',
                        'postalCode': 'postal_code',
                        'lastSeen': 'last_seen'
                        }, inplace=True)
```

```python
autos.head()
```

Changed the column names from camelcase to snakecase, and some of the column names were changed to readability.

## Initial Exploration and Cleaning

Let's do some basic data exploration to determine what other cleaning tasks need to be done. Initially we will look for: 
- Text columns where all or almost all values are the same. These can often be dropped as they don't have useful information for analysis. 
- Examples of numeric data stored as text which can be cleaned and converted.

```python
autos.describe(include='all')
```

Looking at the description of the columns, there are candidates for drops (seller, offer_type, abtest, unrepaired_damage, nr_of_pictures).

Price and odometer are stored as text, should be converted to numeric.

```python
autos["price"] = (autos["price"]
                  .str.replace('$', '')
                  .str.replace(",", "")
                  .astype(int)
                 )

autos["odometer"] = (autos["odometer"]
                    .str.replace('km', '')
                    .str.replace(",", "")
                    .astype(int)
                    )
                    
autos.rename(columns = {"odometer": "odometer_km"}, 
            inplace=True
            )
```

## Dealing with Incorrect Registration Year Data

We have some interesting numbers for min and max: min of 1000 was made many hundreds of years before the model t first rolled down the factory, and max of 9999 is many thousands of years into the future. We'll keep up to 2016, and we have some vehicles < 1911, so we'll take those out

```python
autos['registration_year'].nsmallest(25)
                                    
```

```python
autos = autos[autos["registration_year"].between(1911,2016)]
```

```python
autos['registration_year'].value_counts(normalize=True)
```

Looks like most cars are registered in the past 10 years. Cars with model year 2000 was number 1.

## Exploring Price by Brand

```python
brand_counts = autos["brand"].value_counts(normalize=True)
top_brands = brand_counts[brand_counts > 0.01].index
print(top_brands)
```

German brands take the top 5, takes more than 50% of total brands listed. Let's choose brand with at least 1%.

```python
brand_mean_prices = {}

for brand in top_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

Highest prices go to audi, bmw, and mercedes_benz, lowest prices to fiat, renault, and opel

## Storing Aggregate Data in a DataFrame

```python
bmp_series = pd.Series(brand_mean_prices)
pd.DataFrame(bmp_series, columns=["mean_mileage"])


brand_mean_mileage = {}

for brand in top_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_mileage = brand_only["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending=False)

brand_info = pd.DataFrame(mean_mileage,columns=['mean_mileage'])
brand_info
```

```python

brand_info["mean_price"] = mean_prices
brand_info
```


The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
